```python
from stable_baselines3 import PPO
from StarCraft2Env import StarCraft2Env
```

```python
env = StarCraft2Env()
```

```python
model = PPO.load('./models/1647915989/2880000.zip', env, verbose=1)
```

```python
obs = env.reset()
done = False
iters = 0
while not done:
    iters += 1
    action,_status = model.predict(obs)
    obs, reward, done, info = env.step(action)
    print(f'iters:{iters},reward:{reward}, action:{action}')
```
